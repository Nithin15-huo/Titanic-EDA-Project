### 1.Handle Missing Data

Handling missing values prevents inaccurate analysis and ensures reliable insights.

#### Step 1: Identify Missing Values

Identify which columns have missing or null values.

```python
df.isnull().sum()
```

#### Step 2: Apply Appropriate Strategies

Age is a numerical variable with outliers, so the median was used instead of the mean.

```python
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

Embarked and Embark_town are categorical variables with very few missing values; hence mode imputation was used.

```python
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

Deck contains a large number of missing values, so instead of removing the column, a placeholder category 'Unknown' was assigned.

```python
df['deck'].fillna('Unknown', inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df
```

#### SUMMARY:

Missing values were identified in the age, embarked, deck, and embark_town columns.
Median imputation was used for age, while mode was applied to embarked and embark_town.
Due to high missingness in the deck column, missing values were replaced with 'Unknown'
to avoid introducing bias. All missing values were successfully handled.

#### PHASE-2  SUMMARY:

The initial Titanic dataset contained several data quality issues that required attention before analysis. Missing values were observed in key columns such as age, embarked, deck, and embark_town, with the deck column having a large proportion of missing entries. Additionally, duplicate records were identified and removed to prevent skewed results.

Missing values were handled using appropriate strategies based on data type and significance. Numerical missing values in age were imputed using the median to avoid the impact of outliers. Categorical missing values in embarked and embark_town were filled using the mode, while missing values in deck were replaced with the most frequent category to preserve structural information. Duplicate rows were removed, and categorical inconsistencies such as extra spaces and formatting issues were resolved. Repeated text columns were also converted to categorical data types to improve efficiency and clarity.

Non-visual bivariate analysis revealed meaningful relationships between variables. Survival was strongly influenced by gender, passenger class, and fare, with females and first-class passengers showing higher survival rates. Fare was positively associated with survival, while age alone showed limited influence. Family-related variables such as sibsp and parch displayed a positive relationship, indicating that passengers often traveled in family groups.

Visualization analysis revealed clear patterns in passenger demographics and survival outcomes. Univariate plots showed most passengers were aged 20–40, while fare distribution was highly right-skewed. Bivariate analysis indicated that females and higher-fare passengers had better survival rates. Survivors were generally slightly younger than non-survivors. Multivariate visualizations highlighted how passenger class, embarkation port, and passenger type jointly influenced survival and travel patterns. Overall, visual analysis provided meaningful insights into factors affecting passenger outcomes.
